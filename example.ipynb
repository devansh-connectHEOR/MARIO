```python
from langchain_community.vectorstores import Chroma

vector_store = Chroma(
    collection_name="markdown_docs",
    embedding_function=embedding,
    persist_directory=str(test_outfiles / "chroma_vectorstore"),
)
```

```python
vector_store.add_documents(mkd_docs)
```

```python
vector_store.similarity_search("Network", filter = {"doc_type": "TSD"})
```

```python
from langchain_classic.retrievers import SelfQueryRetriever
from langchain_classic.chains.query_constructor.base import AttributeInfo
from langchain_community.retrievers import BM25Retriever
from langchain_classic.retrievers import EnsembleRetriever
```

```python
metadata_field_info = [
    AttributeInfo(
        name="doc_type",
        description="Whether the document is a TSD (Technical Support Document) or a TA (Technical Assessment).",
        type="string",
    )]

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

retriever = SelfQueryRetriever.from_llm(
    llm, 
    vector_store,  # Your FAISS object
    "NICE technical document database",  # A description of the data in the vector store
    metadata_field_info, 
    verbose=True
)
```

```python
retriever.invoke("Find documents about Network that are TSDs.")
```

```python
retriever.invoke("Find documents about Network that are TAs.")
```

```python
bm25 = BM25Retriever.from_documents(mkd_docs, search_kwargs={"k": 5})
```

```python
bm25.invoke("Parkinson network")
```

```python
ensemble_retriever = EnsembleRetriever(
    retrievers=[retriever, bm25],
    weights=[0.5, 0.5]
)
```

```python
ensemble_retriever.invoke("Documents about Parkinson network that are TAs")
```

```python
from pydantic import BaseModel, Field

class doc_type(BaseModel):
    doc_type: str = Field(description="Whether the document is a TSD (Technical Support Document) or a TA (Technical Assessment).")

llm = ChatOpenAI(model="gpt-4o", temperature=0)


structured_llm = llm.with_structured_output(doc_type)
```

```python
response = structured_llm.invoke("TSD documents about parkinson's network")
```

```python
retriever = vector_store.as_retriever(search_kwargs={"k": 5})
```

```python
query = "Documents about Parkinson network"
doc_type = structured_llm.invoke(query).doc_type

filtered_mkd_docs = [d for d in mkd_docs if d.metadata.get("doc_type") == doc_type]
len(filtered_mkd_docs), doc_type
```

```python
if filtered_mkd_docs:
    print("There are documents of the requested type.")
else:
    print("There are no documents of the requested type.")
```

```python
retriever = vector_store.as_retriever(search_kwargs={"k": 5, "filter": {"doc_type": doc_type}})
retrievers = [retriever, BM25Retriever.from_documents(filtered_mkd_docs, search_kwargs={"k": 5})] if filtered_mkd_docs else [retriever]
weights = [0.5, 0.5] if filtered_mkd_docs else [1.0]
print(retrievers)
ensemble_retriever = EnsembleRetriever(
    retrievers=retrievers,
    weights=weights
)
ensemble_retriever.invoke(query)
```
